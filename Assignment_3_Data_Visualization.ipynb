```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets

# import some data to play with
iris = datasets.load_iris()
iris
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
X = pd.DataFrame(iris.data)
Y = pd.DataFrame(iris.target)
```

```python
X.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
X
```

```python
print(X.shape)
print(Y.shape)
```

```python
#Line Plot 

sns.lineplot(x='sepal_length',y='sepal_width',data=X)
```

```python
#Scatter Plot

sns.scatterplot(x='sepal_length',y='petal_length',data=X)
```

```python
#Bar graph

sns.barplot(x='sepal_length', y='sepal_width', data=X)
```

```python
#Histogram
plt.hist(X['sepal_length'], histtype='bar',rwidth=0.9)
plt.title('Histogram of Sepal Length')
```

```python
#Count plot
sns.countplot(iris.target)
```

```python
#Heat Map
X.corr()
sns.heatmap(X.corr(),annot=True)
```

```python
#Boxplot
sns.boxplot(x = iris.target, y = X['petal_length'])
```

```python
#PairPlot
sns.pairplot(X)
```

```python
#Joinplot
sns.jointplot(x='petal_length',y='petal_width',data=X)
```

```python
#Kernal Density Plot

sns.kdeplot(data=X, y="petal_length")
```
